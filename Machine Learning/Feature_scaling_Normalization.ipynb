```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
```

```python
NBA_FILE_PATH = 'dataset/NBA_player_of_the_week.csv'     # 데이터 파일 경로 설정
nba_player_of_the_week_df = pd.read_csv(NBA_FILE_PATH)     # 데이터 pandas 로 불러오기
nba_player_of_the_week_df
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM', 'Weight KG', 'Age']]     # Normalize 해줄 column 선택해서 가져오기
```

```python
scaler = preprocessing.MinMaxScaler()     # MinMax Scaler 설정
```

```python
normalized_data = scaler.fit_transform(height_weight_age_df)     # Normalization
```

```python
normalized_df = pd.DataFrame(normalized_data, columns = ['Height', 'Weight', 'Age'])     # Normalizaed 데이터 를 DataFrame 으로 변환
```
