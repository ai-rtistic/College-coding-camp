```python
from sklearn import preprocessing

import pandas as pd
import numpy as np
```

```python
NBA_FILE_PATH = 'dataset/NBA_player_of_the_week.csv'     # 데이터 파일 경로 설정
pd.set_option('display.float_format', lambda x: '%.5f' % x)     # 소수점 5번째 자리까지만 출력되도록 설정
nba_player_of_the_week_df = pd.read_csv(NBA_FILE_PATH)     # 데이터 pandas 로 불러오기
nba_player_of_the_week_df
```

```python
height_weight_age_df = nba_player_of_the_week_df[['Height CM', 'Weight KG', 'Age']]     # Normalize 해줄 column 선택해서 가져오기

scaler = preprocessing.StandardScaler()     # Standardized 설정
standardized_data = scaler.fit_transform(height_weight_age_df)     # Standardized

standardized_df = pd.DataFrame(standardized_data, columns = ['Height', 'Weight', 'Age'])     # Standardized 데이터 를 DataFrame 으로 변환

standardized_df.describe()     # 평균(mean)이 0, 표준 편차(std)가 1이 된 것을 확인할수 있음
```
