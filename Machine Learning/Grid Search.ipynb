```python
# 최적의 하이퍼 파라미터를 찾기 위해 Grid Search

from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
from math import sqrt

import numpy as np
import pandas as pd
```

```python
ADMISSION_FILE_PATH ='dataset/admission_data.csv'
admission_df = pd.read_csv(ADMISSION_FILE_PATH)

X = admission_df.drop(['Chance of Admit '], axis=1)

polynomial_transformer = PolynomialFeatures(2)     # 2 차식 변형기
polynomial_features = polynomial_transformer.fit_transform(X.values)
features = polynomial_transformer.get_feature_names(X.columns)

X = pd.DataFrame(polynomial_features, columns = features)
y = admission_df[['Chance of Admit ']]
```

```python
hyper_parameter = {
    'alpha' : [0.01, 0.1, 1, 10],
    'max_iter' : [100, 500, 1000, 1500, 2000]
}     # 비교 해볼 하이퍼 파라미터를 dictionary 형태로 넣어줌   , 보통 default 주변의 값들을 넣어줌
```

```python
lasso_model = Lasso()
```

```python
hyper_parameter_tuner = GridSearchCV(lasso_model, hyper_parameter, cv=5)     # Grid Search
# GridSearchCV(모델, 하이퍼 파라미터 딕셔너리, , cv= 몇 겹 검증 할건지)
hyper_parameter_tuner.fit(X, y)
```

```python
hyper_parameter_tuner.best_params_     # 최적의 하이퍼 파라미터
```
